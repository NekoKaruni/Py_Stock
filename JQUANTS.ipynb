```python
#上場銘柄市場別
import requests
import json
import pandas as pd

# J-Quants API トークン
ID_TOKEN = userdata.get('ID_TOKEN')  # ユーザーのトークンを取得

# APIリクエストのヘッダー
headers = {'Authorization': f'Bearer {ID_TOKEN}'}

# APIから上場銘柄一覧を取得
url = "https://api.jquants.com/v1/listed/info"
response = requests.get(url, headers=headers)

# APIレスポンスをJSON形式で取得
if response.status_code == 200:
    data = response.json()
    info_list = data.get("info", [])  # "info" のリストを取得
else:
    print(f"APIリクエスト失敗: {response.status_code}")
    exit()

# 市場ごとに分類
markets = ["プライム", "スタンダード", "グロース"]

for market in markets:
    # 指定市場のデータを抽出
    filtered_data = [entry for entry in info_list if entry["MarketCodeName"] == market]

    # データがある場合のみ CSV に保存
    if filtered_data:
        df = pd.DataFrame(filtered_data)
        filename = f"{market}_market_data.csv"
        df.to_csv(filename, index=False, encoding="utf-8-sig")
        print(f"CSVファイル '{filename}' を作成しました！")
    else:
        print(f"{market} のデータがありません。")
```
